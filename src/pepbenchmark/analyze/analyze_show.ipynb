```python
from pepbenchmark.analyze.data_analyzer import ClassDataAnalyzer, RegressDataAnalyzer, analyze_class_dataset, analyze_regression_dataset
from pepbenchmark.metadata import DATASET_MAP
from pepbenchmark.utils.logging import get_logger

logger = get_logger()

dataset_names = [
        "bbp"]
for dataset_name in dataset_names:
    base_data_dir = DATASET_MAP.get(dataset_name, {}).get("path", None)
    try:
        stats = analyze_class_dataset(
            dataset_name=dataset_name,
            base_data_dir=base_data_dir,
            similarity_threshold=0.9,
            processes=8,
            print_summary=True,
            enable_visualization=True,
            save_plots=True,
            plot_save_dir=None
        )
        print(f"\nAnalysis completed successfully for {dataset_name}!")
    except Exception as e:
        logger.error(f"Analysis failed for {dataset_name}: {e}")
        print(f"Analysis failed for {dataset_name}: {e}") 
```
